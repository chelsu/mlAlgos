# Building the KNN classifier by finding best value of K

```python
dataNorm.shape
```

```python
X = dataNorm.drop('Outcome',axis=1).values
y = dataNorm['Outcome'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42, stratify=y)
```

```python
#import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

#Setup arrays to store training and test accuracies
def determineKUsing(distanceMetric):
    neighbors = np.arange(1,20)
    train_accuracy =np.empty(len(neighbors))
    test_accuracy = np.empty(len(neighbors))

    for i,k in enumerate(neighbors):
        #Setup a knn classifier with k neighbors
        knn = KNeighborsClassifier(n_neighbors=k, metric=distanceMetric)

        #Fit the model
        knn.fit(X_train, y_train)

        #Compute accuracy on the training set
        train_accuracy[i] = knn.score(X_train, y_train)
        #print("Training accuracy for",i,"is", train_accuracy[i])

        #Compute accuracy on the test set
        test_accuracy[i] = knn.score(X_test, y_test) 
        test_accuracy[i]
        print("Accuracy for",i+1,"is", test_accuracy[i])
        y_pred = knn.predict(X_test)
        print("F1 score is", f1_score(y_test, y_pred, average=None))
    plt.title('k-NN Varying number of neighbors')
    plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
    plt.plot(neighbors, train_accuracy, label='Training accuracy')
    plt.xticks(range(1, 20))
    plt.legend()
    plt.xlabel('Number of neighbors')
    plt.ylabel('Accuracy')
    plt.show()
```

```python
determineKUsing('minkowski')
```

```python
determineKUsing('manhattan')
```

```python
determineKUsing('hamming')
```

```python
determineKUsing('jaccard')
```

### K = 5 

```python
#Setup a knn classifier with k neighbors
knn = KNeighborsClassifier(n_neighbors=5)
#Fit the model
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

# Model performance Analysis 

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = knn.predict(X_test)
confusion_matrix(y_test,y_pred)
```
